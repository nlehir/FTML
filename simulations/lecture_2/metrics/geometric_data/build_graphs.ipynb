```python
threshold = 1.5

for i, j in [(i, j) for i in range(0, x.shape[0]) for j in range(0, x.shape[0])]:
    # we look points a and b
    x_a = x[i]
    y_a = y[i]
    x_b = x[j]
    y_b = y[j]

    # EDIT THE DISTANCE HERE
    distance = 1
    if distance <= threshold:
        if i is not j:
            connectpoints(x, y, i, j)

plt.show()
```
